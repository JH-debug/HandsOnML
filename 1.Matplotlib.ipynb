## 1. Matplotlib

Matplotlib은 파이썬에서 메틀렙의 형식과 흡사한 방식으로 그래프를 그리는 패키지입니다. 하지만 우리는 메틀렙도 써본 적이 없지요. 저런. 그냥 그림을 그릴 때 사용하는 패키지로 알고 계시면 될 것 같습니다. 기본적인 데이터셋이나, numpy의 array데이터를 통해 그림을 그릴 수 있습니다.

```python
import matplotlib.pyplot as plt
```

matplotlib을 호출하라는 명령입니다. 일반적으로 matplotlib의 하위 function인 pyplot까지 호출하여 plt alias에 할당합니다. 
책도 그렇고, 대부분 pyplot까지 호출하는 것을 기본으로 하기에, 이 글에서도 같은 양식을 따릅니다.

```python
%matplotlib inline
```

이 부분은 그림을 이 주피터 노트북 안에 그려달라는, 주피터 노트북의 자체적인 매직 펑션입니다 (주피터 환경을 벗어나면 안돌아간다는 이야기입니다). 가끔 코드를 입력해도 이상한 주소만 반환하는 경우가 있는데, 이를 예방하기 위함입니다. 멧플롯립 1.5 이상에서는 자동으로 그린다는데, 저는 심심찮게 안그래서 이것까지 쓰고 시작합니다. 

#### 직접 plotting

```python
import numpy as np
x=np.linspace(0,10,101) # 0 ~ 10을 101개로 쪼개어 x에 할당
plt.plot(x,np.sin(x))
```

plt.plot(x,y)를 통해, 기본적인 2차원 그림을 그릴 수 있습니다. 이 경우에는 linspace함수를 통해 x에는 0,0.1 ... 9.8, 9.9, 10을 할당하였고, y에는 그 값에 sin을 취한 값을 할당하였습니다. 즉 (0, sin0), (0.1, sin0.1) ... (9.9, sin9.9), (10, sin10)을 모두 선으로 이은 그래프라고 생각하시면 되겠습니다. 만일 plt.plot안에 하나의 array만 입력하게 될 경우에는, 자동으로 y값으로 인식하여 그림을 그리게 됩니다.

```python
plt.plot(x,np.sin(x))
plt.xlim(-1,11) # x축 범위
plt.ylim(-1.5,1.5) # y축 범위
#plt.axis([-1,11,-1.5,1.5])
```

plt.xlim, ylim을 통해 축의 범위를 설정해 줄 수 있습니다. 같은 작업을 plt.axis([a,b,c,d])를 통해 설정할 수도 있습니다. 이 때 a:b가 x축의 범위, c:d가 y축의 범위가 됩니다. 

```python
plt.plot(x,np.sin(x),'b-o',label='sin',linewidth = 2) 
plt.plot(x,np.cos(x),'r:',label='cos')
plt.title('sine-cosine curve')
plt.legend(loc = 'lower left')
```

그 외에, plt.plot(x,y, 이 자리!) 에 그래프의 스타일을 지정해 줄 수 있습니다. 'b-o'와 'r:'은 모두 그래프의 스타일을 지정해 준 부분입니다. b와 r은 색을 blue, red로 정하라는 뜻이고 -o는 큰 점들을 선으로 이어라, :는 점선으로 그어라 하는 명령입니다. 색깔은 대충 이거 있겠지 싶은 애들 첫 글자 넣으면 있습니다. 그 외에 matplotlib에 저장된 스타일에 대해서는 https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html 을 참고하시면 되겠습니다. 

혹은 linewidth = 2 와 같은 옵션으로 두께를, c = 'red'와 같은 옵션으로 색깔을, linestyle = '--'와 같은 옵션으로 선의 종류를 지정해 줄 수도 있습니다. 

한편 plt.legend(loc = 'best') 이 부분은 범례를 그리라는 명령입니다. loc은 위치를 지정해주는 옵션입니다. 지정하지 않으면 알아서 best로 정해지며, 알아서 좋은 자리에 넣어줍니다. 그 외에 upper, left, right, lower등을 조합하여 사용하실 수도 있습니다. 
단 legend를 사용할 때는 label을 지정해주어야 합니다. 그래야 범례가 이게 뭔지 나타낼 수 있겠죠?

#### axes와 figure를 사용한 plotting

한편, 보다 구체적인 그래프를 위해서, fig와 ax라는 객체를 사용해 줄 수 있습니다. 이를 위해서는 plt.subplot이라는 함수가 필요합니다.

이 때 fig는 전체 그림, ax는 각각의 그래프를 뜻한다고 생각하시면 좀 이해가 쉽습니다. 자세한 내용은 뒤에서 살펴 볼게요


```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
S = np.sin(X)
C = np.cos(X)

fig, ax = plt.subplots()

ax.plot(X, S, 'r-', linewidth = 4, label = "sine")
ax.plot(X, C, label = 'cosine')
ax.legend()
```

앞서 말했듯이, 그래프 각각은 ax라는 객체를 통해 호출합니다. 앞에서 한 것과 똑같이 plot 함수를 통해 간단하게 그려줄 수 있으며, 세부 옵션도 같은 방식으로 지정할 수 있습니다.

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
S, C = np.sin(X), np.cos(X)

fig, ax = plt.subplots()

ax.plot(X, C, c="blue", linewidth=4, linestyle="--")
ax.plot(X, S, c="red", linewidth=1, linestyle="-")
ax.set_xlim(X.min() * 1.5, X.max() * 1.5)
ax.set_ylim(C.min() * 1.5, C.max() * 1.5)

ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_yticks([-1, 0, +1])
ax.set_xticklabels(["-pi", "-pi/2", "0", "+pi/2", "+pi"])
ax.set_yticklabels(["-1", "0", "+1"])
ax.set_title("sine and cosine graph")
```

한편, ax의 경우는 xlim을 지정할 때 set_을 앞에 붙여줍니다. 

비슷한 방식으로 set_xticks, set_yticks를 통해 축의 눈금값을 지정해줄 수 있으며, set_xtickslabels / ytickslabels를 통해 해당 눈금값을 단순한 숫자가 아니라 문자열로 지정해 줄 수 있습니다. 

set_title은 제목을 지정합니다. 

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
S, C = np.sin(X), np.cos(X)

fig, ax = plt.subplots()

ax.plot(X, C, c="blue", linewidth=4, linestyle="--")
ax.plot(X, S, c="red", linewidth=1, linestyle="-")
ax.set_xlim(X.min() * 1.5, X.max() * 1.5)
ax.set_ylim(C.min() * 1.5, C.max() * 1.5)

ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_yticks([-1, 0, +1])
ax.set_xticklabels(["-pi", "-pi/2", "0", "+pi/2", "+pi"])
ax.set_yticklabels(["-1", "0", "+1"])

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('axes',0))
```

ax.spine[]을 통해 각각의 축을 만져줄 수 있습니다. 그래프에는 'top','bottom','left','right' 총 4개의 축이 있으며, set_color()를 통해 색깔을, set_position()을 통해 그 위치를 정해줄 수 있습니다. set_position('data',0)을 적어주면 데이터의 가운데로 이동하며, set_position('axes',0)을 적어주면 원래의 위치에 남습니다. 

```python
X = [0.25, 0.5, 0.75, 1, 0]
Y = [0.5, 0, 0.5, 1, 1]

fig, axes = plt.subplots(1, 2)

fig.suptitle("Figure")

axes[0].set_title("Ax 1")
axes[0].scatter(X, Y, c="blue")

axes[1].set_title("Ax 2")
axes[1].scatter(1, 1, c="red")

plt.show()
```

한편, subplots(1,2) 이런 식으로 1행 2열 짜리 그림을 그려줄 수도 있습니다. R을 쓸 줄 아시는 분은 par(mfrow = c(1,2)) 이 옵션을 생각하시면 이해가 편하실 겁니다. 이 경우에, 각각의 그래프를 나타내는 ax는 객채간의 구분을 위해서 axes를 씁니다.

다들 아시겠지만, python의 indexing이 0부터 시작하기 때문에, 첫 그림은 axes[0].하면서 나가고, 그 다음은 axes[1]. 하면서 나가게 됩니다. 

이번에는 plot자리에 scatter가 쓰였지요? 이는 점을 찍어주는 함수입니다. x,y좌표순으로 산점도를 그려주지요. 

제가 앞에서 figure는 전체 그림이라고 했지요?ax.set_title()이 각각의 그래프의 제목이라면, figure.suptitle()은 전체 그림의 제목을 지정해줄 수 있습니다.

```python
import pandas as pd

name = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
score = [9.38, 8.09, 11.32, 5.41, 11.33]

df = pd.DataFrame({"name":name, "score":score})
df

fig, ax = plt.subplots()
ax.bar(df["name"], df["score"], 0.35, color = 'r')
```

ax.bar(x,heigth,width, color)는 막대 그래프를 그려줍니다. 각각 x값, x의 높이, 너비, 색깔을 설정해줄 수 있습니다. 너비의 기본값은 0.8입니다. 

보다 자세한 스타일은 https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.bar.html 이곳을 참조하세요

```python
import pandas as pd
bn=np.random.normal(0,1,size=1000)

plt.hist(bn,bins=5,density=True,cumulative=True, orientation='horizontal',rwidth=0.8, # 1.0일 경우, 꽉 채움 작아질수록 간격이 생김
        )
```

plt.hist(x,bins, density, cumulative, orientation, rwidth, color)는 히스토그램을 생성합니다. bins 는 hist의 그룹의 개수. density는 퍼센트로 표현할지의 여부, cumulatvie는 누적도수를 사용할지 여부를 지정합니다. 둘 다 디폴트는 False입니다. orientation은 방향을 지정하는 옵션으로 'horizontal' 과 'vertical'을 지정할 수 있으며 디폴트는 vertical입니다. rwidth는 너비를 지정하는 옵션으로, 디폴트는 1으로 꽉 찬 그래프이며, 그보다 작아질 수록 그룹 간의 간격이 넓어집니다. color는 색이구요.
